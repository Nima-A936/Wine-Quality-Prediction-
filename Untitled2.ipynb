```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import os
from matplotlib import pyplot as plt
import cv2
```

```python
df = pd.read_csv(
    os.path.join("winequality-red.csv")
)
```

```python
df
```

```python
x = df.drop("quality" , axis=1)
y = df['quality']
labels = df['quality'].unique()
scaled = df["quality"].value_counts()
print(f'labels : {labels} ')
print(scaled)
```

```python
ex = {3:"Bad" ,4:"Bad" ,5:"Mid" ,6:"Fine" ,7:"Fine" ,8:"Fine"}
df["Expression"] = df["quality"].map(ex)
df
```

```python
plt.figure(figsize=(5,5))
plt.style.use("dark_background")
fig , ax = plt.subplots(nrows=1 , ncols=1)
colors = []
for label in labels:
    if label > 5:
        colors.append("green")
    elif label == 5:
        colors.append("gold")
    elif label < 5:
        colors.append("red")      
plot = ax.bar(
    labels ,scaled , color=colors , edgecolor='black',
    linewidth=0.2 , 
)
ax.set_title("Red-Wine-Quality" , fontname="Arial" , fontstyle="italic")
ax.set_xlabel("Qualities" , fontname="Arial" , fontstyle="italic" , size = 12)
plt.xticks(labels , fontname='Arial' , fontstyle='italic' , size=12)
plt.bar_label(plot , fontname='Arial' , fontstyle='italic' , size=12)
plt.grid(False)
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
axs = fig.subplots(nrows=3 , ncols=3)
axs = axs.flatten()
colors=["red","red","red","blue","blue","blue","gold","gold","gold"]
variables = ["fixed acidity","volatile acidity","citric acid","pH","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density"]
for ax,var,col in zip(axs,variables,colors):
    ax.scatter(x=df["alcohol"] , y=df[var] , color=col , s=15)
    ax.set_xlabel("Alcohol")
    ax.set_ylabel(var.title())
    plt.grid(False)
plt.show()    
```
